```python
import turtle

screen = turtle.Screen()
# screen.bgcolor("blue")

t = turtle.Turtle()
t.speed(200)
# draw sun
t.penup()
t.pensize(5)
t.color('orange','yellow')
t.goto(200,150)
t.pendown()


t.begin_fill()
t.circle(50)
t.end_fill()

for i in range(18):
    t.right(90)
    t.fd(40)
    t.bk(40)
    t.lt(90)
    t.circle(50,20)

# draw house
t.penup()
t.goto(-200,-150)
t.pendown()
# draw door
t.fillcolor('cyan')
t.begin_fill()
t.forward(50)
t.left(90)
t.forward(100)
t.left(90)
t.forward(50)
t.left(90)
t.forward(100)
t.end_fill()

# t.fillcolor('red')
# t.begin_fill()
t.right(90)
t.forward(50)
t.right(90)
t.forward(200)
t.right(90)
t.forward(150)
t.right(90)
t.forward(200)
t.right(90)
t.forward(150)
t.right(90)
t.forward(200)
# t.end_fill()

#draw roof and house's body

t.right(30)
t.forward(150)
t.right(120)
t.forward(150)
t.left(90)
t.forward(200)
t.right(120)
t.forward(200)
t.right(60)
t.forward(200)
t.right(120)
t.forward(200)
t.left(30)
t.forward(150)
t.right(90)
t.forward(200)
t.right(90)
t.forward(150)

# draw windows
t.penup()
t.goto(-50,-50)
t.pendown()
t.fillcolor('brown')
t.begin_fill()
t.left(90)
t.forward(100)
t.left(60)
t.forward(70)
t.left(120)
t.forward(50)
t.left(60)
t.forward(70)
t.left(180)
t.forward(70)
t.left(120)
t.forward(50)
t.left(60)
t.forward(70)
t.end_fill()

# draw tree
t.penup()
t.color('black','brown')
t.goto(200,-100)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
t.fd(50)
t.lt(90)
t.fd(20)
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(20)
t.end_fill()

t.color('black','green')
t.fillcolor("green")
t.begin_fill()
t.fd(30)
t.right(120)
t.fd(80)
t.right(120)
t.fd(80)
t.right(120)
t.fd(80)
t.end_fill()

t.penup()
t.color('black','green')
t.goto(215,-10)
t.pendown()

t.fillcolor("green")
t.begin_fill()
t.lt(120)
t.fd(60)
t.right(120)
t.fd(70)
t.right(120)
t.fd(70)
t.right(120)
t.fd(70)
t.end_fill()

t.penup()
t.color('black','green')
t.goto(235,-20)
t.pendown()
t.fillcolor("green")
t.begin_fill()
t.rt(120)
t.fd(50)
t.right(120)
t.fd(50)
t.right(120)
t.fd(50)
t.right(120)
t.fd(50)
t.end_fill()
```
