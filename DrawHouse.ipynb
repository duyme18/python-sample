```python
import turtle

t = turtle.Turtle()

degrees = 90

# draw body
t.penup()
t.goto(-200, -100)
t.pendown()

t.fillcolor("pink")
t.begin_fill()
t.forward(450)
t.left(degrees)
t.forward(300)
t.left(degrees)
t.forward(450)
t.left(degrees)
t.forward(300)
t.end_fill()

# draw door
widthDoor = 80
heightDoor = 150

t.penup()
t.goto(100, -100)
t.pendown()

t.fillcolor('blue')
t.begin_fill()
t.left(180)
t.forward(heightDoor)
t.right(degrees)
t.forward(widthDoor)
t.right(degrees)
t.forward(heightDoor)
t.right(degrees)
t.forward(widthDoor)
t.end_fill()

# draw window
t.penup()
t.goto(-150,150)
t.pendown()

t.fillcolor("red")
t.begin_fill()
t.left(90)
t.forward(60)
t.left(90)
t.forward(100)
t.left(90)
t.forward(60)
t.left(90)
t.forward(100)
t.end_fill()

# draw roof
t.penup()
t.goto(-200,200)
t.pendown()

t.fillcolor('yellow')
t.begin_fill()
t.right(120)
t.forward(150)
t.right(90)
t.forward(260)
t.end_fill()

####

t.penup()
t.goto(200,200)
t.pendown()
t.fillcolor('silver')
t.begin_fill()
t.left(120)
t.forward(50)
t.left(90)
t.forward(50)
t.left(90)
t.forward(50)
t.end_fill()
t.fillcolor('silver')
t.backward(50)
t.begin_fill()
t.right(90)
t.forward(10)
t.right(90)
t.forward(20)
t.right(90)
t.forward(70)
t.right(90)
t.forward(20)
t.right(90)
t.forward(10)
t.end_fill()
```
